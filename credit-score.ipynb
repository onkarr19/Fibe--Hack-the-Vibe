```python
df = pd.read_csv('/kaggle/input/hack-the-vibe/dataset/cleaned_train_data.csv')
test_df = pd.read_csv('/kaggle/input/hack-the-vibe/dataset/cleaned_test_data.csv')
```

```python
df.set_index('id', inplace=True)
test_df.set_index('id', inplace=True)
```

```python
test_df.isnull().sum().sum()
```

```python
np.isinf(test_df).sum().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.3, random_state=42)
```

```python
import xgboost as xgb
```

```python
# XGBoost Classifier
clf = xgb.XGBClassifier(n_estimators=100, max_depth=7, learning_rate=0.1, subsample=0.8, colsample_bytree=0.7, random_state=42)

# Training model on the training set
clf.fit(X_train, y_train)

# Evaluating the performance of model on the testing set
# score = clf.score(X_test, y_test)
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:,1]

from sklearn.metrics import roc_auc_score
print('AUC-ROC score:', roc_auc_score(y_test, y_prob))
```

```python
X_train, y_train = df.drop('target', axis=1), df['target']
```

```python
X_train
```

```python
X_test = test_df
```

```python
# XGBoost Classifier
clf = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, subsample=0.8, colsample_bytree=0.8, random_state=42, gamma=0.5)

# Training model on the training set
clf.fit(X_train, y_train)

# Evaluating the performance of model on the testing set
# score = clf.score(X_test, y_test)
y_pred = clf.predict(X_test)
y_prob = clf.predict_proba(X_test)[:,1]

# from sklearn.metrics import roc_auc_score
# print('AUC-ROC score:', roc_auc_score(y_test, y_prob))
```

```python
predicted_probability = clf.predict_proba(X_test)[:,1]
```

```python
predicted = y_pred
```

```python
np.round(predicted_probability,2)
```

```python
submission = pd.DataFrame({'predicted':predicted, 
                           'predicted_probability':np.round(predicted_probability,2)})
submission.set_index(test_df.index, inplace=True)
submission
```
